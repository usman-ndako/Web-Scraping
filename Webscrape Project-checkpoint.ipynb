first i imported **requests** in order to be able to download the html code from the url using the "**get**" function

```python
import requests
```

```python
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
data = requests.get(standings_url)
```

running the data.text command will show the html code that was requested

```python
data.text
```

I now imported beautifulsoup in order to look at the downloaded html code and use it to select the part of the code i am intrested in

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data.text)
```

```python
data.text
```

```python
standings = soup.select('table.stats_table')[0]
```

```python
standings
```

Use the find_all() function to find the '**a**' tags in the code because that is where the data is located

```python
links = standings.find_all('a')
```

```python
links = [l.get("href") for l in links]
```

```python
links = [l for l in links if '/squads/' in l]
```

```python
links
```

```python
team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
team_urls
```

Extract stats using requests and pandas

```python
chelsea_url = team_urls[9]
```

```python
data = requests.get(chelsea_url)
```

use pandas to select a specific table from the page using the match keyword

```python
import pandas as pd
```

```python
games = pd.read_html(data.text, match='Scores & Fixtures')
```

```python
chelsea_games = games[0]
```

```python
chelsea_games
```

```python
chelsea_games.drop(chelsea_games.index[29:], inplace=True)
```

```python
chelsea_games.head()
```

Get passing stats with requests and pandas

```python
soup = BeautifulSoup(data.text)
```

```python
links = soup.find_all('a')
```

```python
links = [l.get("href") for l in links]
```

```python
links = [l for l in links if l and 'all_comps/passing/' in l]
```

```python
links
```

```python
data = requests.get(f"https://fbref.com{links[0]}")
```

```python
passing = pd.read_html(data.text, match="Passing")[0]
```

Cleaning and merging scraped data with pandas

```python
passing
```

```python
passing.columns = passing.columns.droplevel()
```

```python
passing
```

```python
passing.drop(passing.iloc[:, 15:24], axis = 1, inplace = True)
```

```python
passing
```

```python
for col in passing.columns:
    print(col)
```

```python
game_data = chelsea_games.merge(passing[['Date','TotDist','PrgDist','Prog','CrsPA','1/3','KP','PPA']], on="Date")
```
